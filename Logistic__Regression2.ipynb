```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/accident_prediction_india.csv")
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df['Accident Severity'].unique()
```

```python
df=df[df['Accident Severity'] != "Fatal"]
df['Accident Severity'] = df['Accident Severity'].str.strip()
df['Accident Severity']=df['Accident Severity'].map({
    'Minor':0, 'Serious':1
})
```

```python
df.head(5)
```

```python
df_new=df[[
    'Number of Vehicles Involved',
    'Number of Casualties',
    'Speed Limit (km/h)',
    	'Driver Age',
   	'Alcohol Involvement',
    'Accident Severity'

]].copy()
df_new['Alcohol Involvement']=df_new['Alcohol Involvement'].map({
    'Yes':1, 'No':0
})
```

```python

x=df_new.iloc[:,:-1]
y=df_new.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred , y_test))
```
